## < NLP on "name" with Word Cloud >

The object column "name" has descriptions of each rentals. By performing Natural Language Processing, we will find out the popularity of words apeared in descriptions, follwed by visualizing it with Word Clud.

```python
# Import packages

import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Load dataset

df = pd.read_csv("~/Desktop/AirBnbNYC2019.csv") 
df.head()
```

```python
# Load necessary libraries to construct Wordcloud

from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Check any na/null values in the column "name"

df['name'].isnull().values.any()
```

```python
# Drop na/null vallues in "name"

df2 = df.dropna(subset=['name'])
```

```python
# Verify there's no na/null values and it's ready for wordcloud construction

df2['name'].isnull().values.any()
```

```python
# Create a combined text file with all words in "name"

text = " ".join(review for review in df2.name)
print ("There are {} words in the combination of all review.".format(len(text)))
```

```python
# Create stopword list:
stopwords = set(STOPWORDS)
stopwords.update(["apt","room","Brooklyn","nyc","Manhattan","Queens",'Apartment','bedroom'])

# Generate a wordcloud image
wc = WordCloud(stopwords=stopwords, background_color="white", width=3000, height=2000).generate(text)

# Display the generated image in the matplotlib way

plt.figure(figsize=(20,10))
plt.imshow(wc)
```

< Name NLP Summary >

- After setting some "Stop words" such as "apt", "NYC", "Bedroom", we detected that there are some popular words to describe the rooms/homes which have appeared in their description more frequently than others;
    1. "Private"
    2. "Cozy"
    3. "Spacious"
    4. "Beautiful"
    5. "Studio" / "heart"
- We can say that these words are the most important & appealing key words for rental owners to state in their descriptions.
- We can assume that potential renters probably seek for rooms where they can feel "private" and "cozy".
